```python
# import necessary libraries
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
# Mount Google Drive
drive.mount('/content/gdrive', force_remount=False)

# Set the path to the 'Masters_data' folder in your Google Drive
base_path = '/content/gdrive/MyDrive/'
folder_name = 'Masters_data'
folder_path = os.path.join(base_path, folder_name)

# define the file path on your Google Drive
file_path = os.path.join(folder_path, 'final_results_eval.json')

# read in the file to a pandas dataframe
df_metrics = pd.read_json(file_path)

# explode the column(s) that contains lists
# replace 'column_name' with the name of your column
# assuming df is your dataframe


# define the file path on your Google Drive for metadata
metadata_file_path = os.path.join(folder_path, 'metadata.json')

# read in the metadata file to a pandas dataframe
df_metadata = pd.read_json(metadata_file_path)

# merge the two dataframes on the dataset_id column
df = pd.merge(df_metrics, df_metadata, left_on='dataset_id', right_on='id', how='inner')

for col in df.columns:
    df = df.explode(col)

# define a list of metrics you want to plot
metrics = ['memory_mean', 'runtime_mean', 'cp_mean', 'sp_mean', 'db_mean', 'silhouette_mean', 
           'calinski_harabasz_mean', 'ari_mean', 'nmi_mean', 'dvi_mean']
# set a style for seaborn
sns.set(style="whitegrid")
folder_path = "/content/gdrive/MyDrive/Masters_data/Plots"
os.makedirs(folder_path, exist_ok=True)
# loop over each metric
for metric in metrics:
    for col in df_metadata.columns.drop(['id', 'file_path']):
        # Create a pivot table
        table = df.pivot_table(values=metric, index=col, columns='model')
        
        # Create a figure and axis
        fig, ax = plt.subplots(figsize=(10, 6))

        # Plot data
        table.plot(kind='line', ax=ax)

        # Customize plot
        ax.set_title(f'Comparison of {metric} for each model by {col}')
        ax.set_ylabel(metric)
        ax.set_xlabel(col)

        # Save the plot as an image file
        plt.savefig(f'{folder_path}/{metric}_comparison_{col}.png')
        plt.close()
```
